## Numpy (numerical python), scipy (scientific python) and math are packages that provide many useful functions. 

Some of these overlap so just be clear about where you are getting them from. They should give the same answer but may have different computation times. This is unlikely to matter for what we're doing here.

```python
import numpy as np # Saves some typing
import scipy
import math
```

```python
print (np.cos(60))
print (scipy.cos(60))
print (math.cos(60))
```

## Numpy is really good at array manipulation

```python
np.arange(3,9) # Will make an array from the first number up until (not including) the last
```

```python
# Can select a subset by making a mask
arr = np.arange(3,9)
arr[arr<6]
```

```python
# Can combine conditions
arr[((arr>4) & (arr<8))]
```

### You can also use numpy to convert from a list...

```python
# Use numpy to convert list to array
mylist = [1, 2, 3]
print (mylist, ' is a ', type(mylist))
myarray = np.asarray(mylist)
print (myarray, 'is a ', type(myarray))
```

## Scipy is really good for stats, Fourier Transform (see microstructure practical), integration, interpolation

```python
# e.g. plot normal distribution
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
# e.g. integrate sin (x) for x = 0 to pi
import scipy.integrate as integrate
integrate.quad(lambda x: np.sin(x), 0, np.pi)
```

```python
# Interpolate missing data
from scipy import interpolate
from scipy.interpolate import CubicSpline

# Define some points
x = np.linspace(0, 4*np.pi, 15)
y = np.sin(x)

plt.close()
plt.plot(x,y, 'bo', label='data')

#x = np.arange(0, 10)
#y = np.exp(-x/3.0)

newx = np.arange(0, 4*np.pi, 0.01)
cs = interpolate.CubicSpline(x, y)

f = interpolate.interp1d(x, y)
plt.plot(newx, f(newx), 'k', label='1D interp')
plt.plot(newx, cs(newx), 'r--', label='cubic spline')
plt.legend()
```

### Or save some computational time if you're working with scalars

```python
import timeit

def f_np():
    return np.sin(10)

def f_math():
    return math.sin(10)
```

```python
# Time taken to do 100000 numpy calculations
timeit.timeit(f_np, number=100000)
```

```python
# Time taken to do 100000 equivalent math calculations
timeit.timeit(f_math, number=100000)
```
